```python
# Get text (<p> scraper?)
# from each link, and throw errors for abnormalities. Store the scraped
# text in a file (journal_data/txt/UCSD).

# Use previous code to generate Hillel mentions graph for UCSD.

# Note: sort pdfs by descending date until 2009

# Observations notes:

# Focus on subpages 2-33 as they share a link pattern.
#________________________________________________________________________


# get links to all subpages procedurally, and store
# the subpages in a list using a json file.
import requests
from bs4 import BeautifulSoup
import json
import re

# Base URL for the UCSD newspaper archives
base_url = "https://library.ucsd.edu/dc/search?f%5Bcollection_sim%5D%5B%5D=UCSD+Guardian&page="

# List to store all the links
all_links = []
subpage_links = []

# Loop through all the pages (2-33) and scrape the links
for page_num in range(1, 34):
    # Construct the URL for the current page
    url = base_url + str(page_num) + "&sort=object_create_dtsi+desc%2C+title_ssi+asc"
    subpage_links.append(url)

with open('subpage_links.json', 'w') as f:
    json.dump(subpage_links, f, indent=4, separators=(",\n", ": "))

# Within each subpage, open the respective subdirectory and
# get the date (YYYY_mm_dd) & the pdf link. Create a separate
# json. file to store the pdf links (ucsd_article_pages.json).

pdf_links = []

# Loop through all the subpage links and scrape the article links
for subpage_url in subpage_links:
    # Make a GET request to the subpage URL
    response = requests.get(subpage_url)

    # Create a BeautifulSoup object from the response HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the article links on the page
    articles = soup.find_all('div', class_='title-link')
print(articles)

    # Loop through all the articles and extract the date and PDF link
for article in articles: 
        try:
    
            # Extract the relative URL for the PDF file
            pdf_url = article.find('a', class_='title-link')['href']
        
            # Construct the absolute URL for the PDF file
            pdf_suffix = re.search(r'/_(\d+)\.', pdf_url).group(1)
            pdf_full_url = f"https://library.ucsd.edu{pdf_url}/_{pdf_suffix}.pdf"

            # Extract the date from the PDF URL
            date_str = pdf_url.split('/')[-2]
            year_str = date_str[:4]
            month_str = date_str[4:6]
            day_str = date_str[6:]
            date = f"{year_str}-{month_str}-{day_str}"

            # Create a dictionary with the date and PDF URL
            pdf_info = {
                'date': date,
                'pdf_url': pdf_full_url
            }

            # Append the dictionary to the list of PDF links
            pdf_links.append(pdf_info)
        except TypeError:
        # Skip articles that do not have a PDF link
            continue

if len(pdf_links) > 0:
    with open('ucsd_article_pages.json', 'w') as f:
        json.dump(pdf_links, f, indent=4)
else:
    print("No PDF links found.")
    print(response.content)


# Close the file after writing
f.close()















#num_pages = 33
#links_to_process = set()
#for i in range(1, num_pages+1):
 #   link = base_url + str(i)
 #   links_to_process.add(link)
## url = base_url.format(i)
   # html_content = requests.get(url).text
   # soup = BeautifulSoup(html_content, "html.parser")
   # links = soup.select('a[href*="page="]')

   # for link in links:
     #   href = link.get('href')
      #  if not href.startswith('http'):
      #      href = base_url.rstrip('/') + '/' + href.lstrip('/')
      #  all_links.append(href)


    # Send a GET request to the URL and get the HTML content
#response = requests.get(url)
#html_content = response.content

    # Parse the HTML content using BeautifulSoup
#soup = BeautifulSoup(html_content, "html.parser")

    # Find all the links in the HTML content that match the pattern
   # links = soup.find_all("a", {"class": "title-link"})
   # for link in links:
        #href = link.get("href")
       # if href.startswith("/dc/"):
            # Construct the full URL for the link
          #  full_url = "https://library.ucsd.edu" + href
          #  all_links.append(full_url)

# Sort the links in descending order
#all_links = sorted(all_links, reverse=True)


# Write the links to a JSON file
#with open("ucsd_newspaper_links.json", "w") as f:
    #json.dump(all_links, f)

```
